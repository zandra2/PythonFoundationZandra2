```python
# run plots in the notebook 
%matplotlib inline
```

```python
url = "http://pbpython.com/extras/sample-salesv2.csv"
```

```python
sales = pd.read_csv(url)
```

```python
sales.head()
```

```python
sales.columns = ['account number', 'name', 'sku', 'category', 'quantity', 'unit_price', 'ext_price', 'date']
```

```python
sales.head()
```

```python
sales.describe()
```

```python
sales[['unit_price','quantity']].plot()
```

```python
sales['unit_price'].hist()
```

```python
sales['category'].value_counts()
```

```python
sales['category'].value_counts().plot(kind='bar')
```

```python
belt_df = sales[['name', 'category', 'quantity', 'unit_price']]
```

```python
belt_df.sort_values(by = 'quantity', ascending = False)
```

```python
#which company sold the most belt
belt_df.head()
```

```python
belt_df = belt_df[belt_df['category'] == 'Belt']
```

```python
belt_df.head()
```

```python
sorted_belts = belt_df.sort_values(by = 'quantity', ascending = False)
```

```python
sorted_belts.head()
```

```python
belts_by_company = sorted_belts.groupby('name').sum()
```

```python
belts_by_company.head(30)
```

```python
belt_df.groupby('name', as_index=False).sum()
```

```python
top10=belt_df.groupby('name', as_index=False).sum()
```

```python
top10
```

```python
my_plot = top10.plot(kind = 'bar', 
                       title = 'Total Belt Sales by Company',
                       x = 'name',
                       y = 'unit_price')
my_plot.set_xlabel('Company Name')
my_plot.set_ylabel('Belts Sold')
```
